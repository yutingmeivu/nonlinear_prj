```python
import numpy as np
from network import NN
from activations import tanh, tanh_prime, tanh_second
from losses import mse, mse_prime, mse_second
from dists import uniform, norm, gen
```

```python
# training data toy example
x_train = np.array([[[0,0]], [[0,1]], [[1,0]], [[1,1]]])
y_train = np.array([[[0]], [[1]], [[1]], [[0]]])
```

### Gradient Descent

```python
for i in range(10):
    net = NN()
    net.load(x_train, y_train)
    net.add((2, 3), 'fc')
    net.add((None), 'activation')
    net.add((3, 1), 'fc')
    net.add((None), 'activation')

    # train
    net.use(mse, mse_prime, mse_second)
    net.activate(tanh, tanh_prime, tanh_second)
    # net.distribution(norm)
    net.fit(1000, 0.001, 'gd', None, None, None, None, 0.01, None, False, True, None, 0.15)
```

### Steepest descent + Secant method

```python
for i in range(10):    
    net = NN()
    net.load(x_train, y_train)
    net.add((2, 3), 'fc')
    net.add((None), 'activation')
    net.add((3, 1), 'fc')
    net.add((None), 'activation')

    # train
    net.use(mse, mse_prime, mse_second)
    net.activate(tanh, tanh_prime, tanh_second)
    net.distribution(gen)
    net.fit(1000, 0.001, 'steepest', 'secant', \
        None, None, 1000, 0.005, None, False, True, 'cosine', 0.15)
```

We can notice that secant method is not stable compared with gd, but it also has certain probability to reach the threshold for stopping(loss < 0.15 here in this case) with small iterations. From gd we notice that after 1000 iterations the loss overall is still bigger than 0.2

Now let's increase the number of experiments, see the frequency for it to reach the threshold and the number of iterations it take

- K = 1000
